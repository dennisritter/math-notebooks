### <a id="general">General</a>

A probability distribution describes how likely different outcomes are for a random variable. It shows which values are possible and how likely each one is to occur.

A **probability density function (PDF)** describes the **relative likelihood** of a **continuous random variable** taking on a specific value. The area under the PDF curve over an interval gives the probability of the variable falling within that interval.

Several probability distributions are particularly important in statistics and data science. These include the **Normal (or Gaussian) distribution**, **Binomial distribution**, **Poisson distribution**, and **Uniform distribution**.

```python
x = np.linspace(-4, 4, 500)
theoretical_normal_distribution = stats.norm.pdf(x, loc=0, scale=1)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(
    x=x,
    y=theoretical_normal_distribution,
    color="red",
    label="Normal Distribution",
    ax=ax,
)
plt.show()
```

### <a id="histogram">Histogram</a>

Histograms are a tool for visualizing distributions through bars. A histogram divides the range of data values into classes/intervals/bins. The height of each bar reporesents the frequency of observations falling into that interval.

#### Applications
* Visualize the shape of a distribution (e.g., symmetric, right-skewed, left-skewed)
* Identify outliers
* Estimate central tendencies and dispersion of the data
* Recognize multimodal distributions

```python
ax = sns.displot(TIPS_DATA, x="tip")
```

### <a id="normal">Normal Distribution</a>

* Always centered on the average value

```python
random_normal_distribution = np.random.normal(0, 1, 1000)
x = np.linspace(-4, 4, 500)
theoretical_normal_distribution = stats.norm.pdf(x, loc=0, scale=1)

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(
    x=x,
    y=theoretical_normal_distribution,
    color="red",
    label="theoretical",
    ax=ax,
)
sns.kdeplot(
    random_normal_distribution,
    color="blue",
    label="random",
    fill=True,
    ax=ax,
)

ax.legend()
plt.tight_layout()
plt.show()
```
